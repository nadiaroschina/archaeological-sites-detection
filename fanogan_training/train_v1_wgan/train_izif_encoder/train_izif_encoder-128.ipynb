# Датасет

```python
transform_crop_128 = transforms.Compose([
    transforms.CenterCrop((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])
```

```python
train_dataset = ArchNegatives(transform=transform_crop_128)
train_loader = DataLoader(train_dataset, batch_size=params['batch_size'], shuffle=True)

anomalies_dataset = ArchPositives(transform=transform_crop_128)
anomalies_loader = DataLoader(anomalies_dataset, batch_size=params['batch_size'], shuffle=False)
```

```python
print(f'train: {len(train_dataset)} images, {len(train_loader)} batches')
print(f'anomalies: {len(anomalies_dataset)} images, {len(anomalies_loader)} batches')
```

```python
fig, axs = plt.subplots(nrows=2, ncols=4)

for i in range(4):
    axs[0, i].axis('off')
    axs[1, i].axis('off')
    
    axs[0, i].imshow((anomalies_dataset[i][0].permute(1, 2, 0) + 1) / 2)
    axs[1, i].imshow((train_dataset[i][0].permute(1, 2, 0) + 1) / 2)
    
fig.tight_layout()
fig.show()
```
