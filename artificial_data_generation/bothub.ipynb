```python
import json
import requests

from io import BytesIO
from PIL import Image

from openai import OpenAI
```

```python
TOKEN = ...
```

```python
client = OpenAI(
    api_key=TOKEN,
    base_url='https://bothub.chat/api/v2/openai/v1'
)
```

```python
params = {
    'model': 'dall-e-3',
    'prompt': 'Satellite image of a desert with an archeological site. View directly from above, as if taken from a plane. The image should cover approximately 100m * 100m.',
    'n': 1,
    'size': '1024x1024',
}

req = client.images.generate(**params)
image_url = json.loads(req.model_dump_json())['data'][0]['url']
image_url
```

```python
img = Image.open(BytesIO(requests.get(image_url).content))
img
```
